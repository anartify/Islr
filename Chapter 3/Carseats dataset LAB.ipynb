```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
CS = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Carseats.csv')
CS.head(5)
```

```python
dCS = pd.get_dummies(CS)
dCS.head(5)
```

```python
reg = smf.ols('Sales~'+'+'.join(dCS.drop(['Sales'], axis=1).columns)+'+Income:Advertising + Price:Age', data=dCS).fit()
reg.summary()
```
